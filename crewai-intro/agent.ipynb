```python
from dotenv import load_dotenv
_ = load_dotenv()
    
from crewai import Crew, Process, Agent, Task

```

```python
planejador_de_viagem = Agent(
    role="Planejador de Viagem",
    goal="Planejar todos os detalhes de uma viagem, incluindo roteiros, hospedagens e atividades.",
    backstory="""
    VocÃª Ã© um especialista em planejamento de viagens, apaixonado por criar experiÃªncias inesquecÃ­veis.
    Seu objetivo Ã© organizar todos os detalhes da viagem de forma eficiente, confortÃ¡vel e divertida,
    garantindo que cada etapa seja bem aproveitada.
    """,
    verbose=True
)
```

```python
ocamentista = Agent(
    role="OrÃ§amentista de viagem",
    goal="Estimar o custo total de uma viagem, considerando transporte, alimentaÃ§Ã£o, hospedagem e lazer",
    backstory= """
    VocÃª Ã© um analista financeiro focado em viagens. Sua missÃ£o Ã© garantir que os custos esteja dentro do
    orÃ§amento, criando estimativas precisas para cada parte da viagem.
    """
)
```

```python
planeja_roteiro =  Task(
    description= """crie um roteiro detalhado para uma viagem na Europa 
    incluindo cidades, atividades e o transporte
    """,
    agent=planejador_de_viagem,
    expected_output="""
    um roteiro com a sequencia de cidades a serem visitadas,
    as principais atividades e o tipo de transporte utilizado
    """
)
```

```python
estima_orcamento = Task(
    description= """
    calcule o orÃ§amento total da viagem levando em consideraÃ§Ã£o
    as cidades, transportes, hospedagens e os lazeres
    """,
    agent = ocamentista,
    expected_output= """
    Uma estimativa de orÃ§amento detalhado com custos aproximados para cada item da viagem
    """
)
```

```python
viagem_crew = Crew(
    agents = [planejador_de_viagem, ocamentista],
    tasks = [planeja_roteiro, estima_orcamento],
    process = Process.sequential
)
```

```python
result = viagem_crew.kickoff()
```
